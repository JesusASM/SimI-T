b) Ahora, realizamos la simulación n-ésima de la suma de cinco variables aleatorias que sigan la distribución triangular y contabilizamos las que excedan la carga.
Calculamos la probabilidad de forma que:

$$ P(X > 1T) = \frac{Excedencias}{Total\ de\ simulaciones} $$

```python
import numpy as np
import matplotlib.pyplot as plt

def Ej5(a,b,c,nb,nE,w):
  e=0
  xT=[]
  for i in range(nE):
    w_5=np.random.triangular(a,c,b,size=nb)
    w_T=np.sum(w_5)
    xT.append(w_T)
    if w_T>w:
      e=e+1
  P=e/nE
  print("La probabilidad es:",P,"\n")
  plt.hist(xT,100)

a=190; b=230; c=210; nb=5; nE=10000; w=1000;

Ej5(a,b,c,nb,nE,w)
```

Para la forma analítica, primero consideramos que de las 52 semanas por 5 días de trabajo a la semana se trabajarán 260 días al año.
Además, el costo por día en que se exceda la carga es de $200.
Podemos plantear la siguiente esperanza diaria:

$$ E(X)= P(X>1000) \$200 + (1-P(X>1000)(\$0)) $$

Y considerando que el evento X que consiste en la suma del peso de las tinas son independientes, podemos expresar la esperanza como:

$$E(X_{1:260})= P(X_1>1000) \$200 + P(X_2>1000) \$200 +...+P(X_{260}>1000) \$200 = P(X>1000) \cdot (260) \cdot \$200=0.9969 \cdot \$52000=\$51838.8$$

Por lo que se espera gastar una cantidad menor a la que se gastaría al adquirir un camión extra.


b)\
Por simulación, utilizaremos el código anterior sólo que modificando la cantidad de experimentos a 260 que son los días, y por cada día que se exceda el peso añadiremos $200 al gasto.


```python
import numpy as np
import matplotlib.pyplot as plt

def Ej5(a,b,c,nb,nE,w):
  e=0
  xT=[]
  for i in range(nE):
    w_5=np.random.triangular(a,c,b,size=nb)
    w_T=np.sum(w_5)
    xT.append(w_T)
    if w_T>w:
      e=e+1
  P=e/nE
  print("La probabilidad es:",P,"\n")
  print("Se excede ",e," veces.\n")
  print("Se gasta ",e*200)
  plt.hist(xT,40)

a=190; b=230; c=210; nb=5; nE=260; w=1000;

Ej5(a,b,c,nb,nE,w)
```

Como podemos observar, al ser pocos los experimentos la probabilidad se aproxima a 1, dando $52000 de gasto, que concuerda con la solución analítica, y la desviación ocurre por la cantidad de experimentos realizados.
Concluyendo, no es conveniente comprar un camión.
