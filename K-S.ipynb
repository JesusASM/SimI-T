```python
import random as rnd
from tabulate import tabulate
import numpy as np
import math

def KS(data,i,lam):
    d=len(data)
    CDOb=[];CDEs=[];CCat=[];FAO=[0]*i;FAOr=[0]*i;FAE=[0]*i;FAEr=[0]*i;DAb=[0]*i;D=0;

    for e in range(i):
        locals()[f'C{e}']=0
        CCat.append(e+1)
    for h in range(d):
        f=data[h]
        for e in range(i):
            if f<((e+1)/i):
                locals()[f'C{e}']+=1
                break
    for e in range(i):
        CDOb.append(locals()[f'C{e}'])
        CDEs.append(((lam ** (e+1))*(math.exp(-lam))/math.factorial(e+1)))
        FAE[0]=CDEs[0]
        FAO[0]=CDOb[0]
        FAOr[0]=FAO[0]/d
        if e>0:
            FAE[e]=(CDEs[e]+FAE[e-1])
            FAEr[e]=FAE[e]/d
            FAO[e]=(CDOb[e]+FAO[e-1])
            FAOr[e]=FAO[e]/d
            DAb[e]=np.abs(FAEr[e]-FAOr[e])
            D=np.max(DAb)

    head=["Categor√≠a","Frec Obs","Frec Ac Obs","Frec Ac Obs Rel","Frec Esp","Frec Ac Esp","Frec Ac Esp Rel","Des Abs"]
    dataf=np.column_stack((CCat,CDOb,FAO,FAOr,CDEs,FAE,FAEr,DAb))
    print(tabulate(dataf, headers=head, tablefmt="github"))

    print("El estadistico es: ",D)


#Ejemplo

data=[0.8797,0.3884,0.6289,0.5999,0.8589,0.9996,0.2415,0.3808,0.9606,0.9848,0.3469,0.7977,0.5844,0.8147,0.6431,0.7387,0.5613,0.0318,0.7401,0.4557,0.1592,0.8536,0.8846,0.3410,0.1492,0.8681,0.5291,0.3188,0.5992,0.9170,0.2204,0.5991,0.5461,0.5739,0.3254,0.0856,0.2258,0.4603,0.5027,0.8376,0.6235,0.3681,0.2088,0.1525,0.2006,0.4820,0.4272,0.6360,0.0954]
KS(data,20,4)
```
