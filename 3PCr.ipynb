```python
from tabulate import tabulate
import numpy as np

def PCr(data):
    d=len(data);Nu=[d]*d;SEC=[];COR=[];SEC.append(" ");COR.append(" ")
    for i in range(d-1):
        Nu[i]=i+1
        if data[i]<data[i+1]:
            SEC.append(1)
        else:
            SEC.append(0)
        if SEC[i]!=SEC[i+1]:
            COR.append(1)
        else:
            COR.append(0)
    Co=np.sum(COR[1:d-1]); mc=((2*d)-1)/3; vc=((16*d)-29)/90; zo=np.abs((Co-mc)/(np.sqrt(vc)))
    head=["Número","Dato","Secuencia","Corrida"]
    dataf=np.column_stack((Nu,data,SEC,COR))
    print(tabulate(dataf, headers=head, tablefmt="github"))
    print("Media: ",mc,"\nVarianza: ",vc,"\nEstadístico Zo: ",zo)

  #Ejemplo
data=[0.8797,0.3884,0.6289,0.5999,0.8589,0.9996,0.2415,0.3808,0.9606,0.9848,0.3469,0.7977,0.5844,0.8147,0.6431,0.7387,0.5613,0.0318,0.7401,0.4557,0.1592,0.8536,0.8846,0.3410,0.1492,0.8681,0.5291,0.3188,0.5992,0.9170,0.2204,0.5991,0.5461,0.5739,0.3254,0.0856,0.2258,0.4603,0.5027,0.8376,0.6235,0.3681,0.2088,0.1525,0.2006,0.4820,0.4272,0.6360,0.0954]
PCr(data)
```
